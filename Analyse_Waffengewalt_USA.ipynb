```python
def split_and_count_m(u):
    h=[]
    u=u.split('||')
    for i in u:
        j=i[3:]
        h.append(j)
    f = h.count('Male')
    return f
```

```python
def split_and_count_f(u):
    h=[]
    u=u.split('||')
    for i in u:
        j=i[3:]
        h.append(j)
    f = h.count('Female')
    return f
```

```python
data_raw['Male']=data_raw.participant_gender.apply(str).apply(split_and_count_m)
```

```python
data_raw['Female']=data_raw.participant_gender.apply(str).apply(split_and_count_f)
```
